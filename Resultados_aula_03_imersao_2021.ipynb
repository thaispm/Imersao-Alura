Desafio 5: Calcular e analisar a correlação entre os g e c. 




```python
#Primeira parte seleção:
gene = dados.loc[:,'g0':'g99']
celula = dados.loc[:,'c0':'c99']
dados_selecionados = pd.concat([gene, celula],axis=1)
dados_selecionados
```

```python
corr = dados_selecionados.corr()
corr_selecionados = corr.loc[:'g99','c0':]
```

```python
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_selecionados, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(15, 10))

# Generate a custom diverging colormap
cmap = sns.cubehelix_palette(as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_selecionados, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```
