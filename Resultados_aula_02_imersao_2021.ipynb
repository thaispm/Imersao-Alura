```python
mapa = {'droga' : 'composto'}
dados.rename(columns=mapa, inplace=True)
```

```python
import seaborn as sns

sns.countplot(x = 'composto', data=dados)
```

```python
dados['composto'].value_counts()
```

```python
dados['composto'].value_counts()[:5]
```

```python
cod_compostos = dados['composto'].value_counts().index[:5]
cod_compostos
```

```python
dados.query('composto in @cod_compostos')
```

```python
sns.set()
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'))
ax.set_title('Top 5 Compostos')
plt.show()
```

```python
#Desafio 01: Reordenar barras
#Desafio 02: Melhorar visualização
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize=(8, 6))

ax = sns.countplot(x = 'composto', data = dados.query('composto in @cod_compostos'), order = dados['composto'].value_counts().index[:5])
ax.set_title('Principais Compostos', fontweight = 'bold', fontsize = '20')
plt.xlabel('Composto', fontweight = 'bold', fontsize = '15')
plt.ylabel('Número de vezes testado', fontweight = 'bold', fontsize = '15')
plt.show()
```
